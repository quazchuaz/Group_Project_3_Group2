```python
import pandas as pd
from pathlib import Path
```

```python
# The path to our CSV file
fire = Path("Resources/fire.csv")

# Read our Crowdfunding data into pandas
fire_df = pd.read_csv(fire)
fire_df.head()
```

```python
fire_df.info()
```

```python
copy_fire_df = fire_df
copy_fire_df.head()
```

```python
copy_fire_df["coordinates"] = copy_fire_df[['lat', 'lon']].astype(str).agg(', '.join, axis=1)
copy_fire_df.head()
```

```python
copy_fire_df = copy_fire_df[['coordinates','lat','lon','rep_date','estarea']]
copy_fire_df.head()
```

```python
copy_fire_df['rep_date'] = pd.to_datetime(copy_fire_df['rep_date'])
copy_fire_df.info()
```

```python
# Group by 'coordinates' and find the first spotted time and last spotted times
grouped_df = copy_fire_df.groupby('coordinates')
copy_fire_df['1st_spotted'] = grouped_df['rep_date'].transform('min')
copy_fire_df['last_spotted'] = grouped_df['rep_date'].transform('max')
# Find the max "estarea" for each 'coordinates' and store it in a new column as "total_area"
copy_fire_df['total_area'] = grouped_df['estarea'].transform('max')
```

```python
copy_fire_df.head()
```

```python
# Create a boolean mask to identify rows where '1st_spotted' and 'last_spotted' are different
mask = copy_fire_df['1st_spotted'] != copy_fire_df['last_spotted']

# Filter the DataFrame to get rows where '1st_spotted' and 'last_spotted' are different
result_df = copy_fire_df[mask]
result_df
```

```python
clean_fire_df = copy_fire_df[['coordinates','lat','lon','1st_spotted','last_spotted','total_area']]
clean_fire_df.head()
```

```python
clean_fire_df.nunique()
```

```python
clean_fire_df.drop_duplicates(inplace=True)
clean_fire_df.head()
```

```python
clean_fire_df.info()
```

```python
clean_fire_df.to_csv('Resources/clean_fire.csv', index=False)
```
